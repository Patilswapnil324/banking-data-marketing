```python
#as you can see from classification report we have good accuracy but our recall value for class 1 is 0.38 & this is the case of 
#class of imbalance

# Over-sampling of minority class observation
# It means we will randomly replicate minority class observation until it becomes actionablely proportional to majority class observation.
```

```python
y.value_counts()
```

```python
#will use oversampling method
```

```python
from sklearn.utils import resample

# separate the maority and minority class observation
data_major = df_train[y == 0.0]
data_minor = df_train[y == 1.0]

# over-sample the minority class observations
data_minor_oversample = resample(data_minor, replace = True, n_samples=27932, random_state = 0)

# finally combine the majority class observation and oversampled minoiry class observation
data_oversampled = pd.concat([data_major, data_minor_oversample])
```

```python
data_oversampled.columns
```

```python
data_oversampled.subscribed.value_counts()
#now we can see class is perfectly balanced
```

```python
# again lets splt our over samoled data into feature and traget variables
X = data_oversampled.drop(['subscribed'],axis=1)
Y = data_oversampled.iloc[:,7]
```

```python
# lets split data into train and test set
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.3, random_state = 0)
```

```python
# model building using random forest
classifier = RandomForestClassifier(n_estimators = 45, criterion = 'entropy', random_state = 0)
classifier.fit(x_train, y_train)
y_pred = classifier.predict(x_test)
```

```python
# print the scores on training and test set

print('Training set score: {:.4f}'.format(classifier.score(x_train, y_train)))

print('Test set score: {:.4f}'.format(classifier.score(x_test, y_test)))
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)  #score is less than both above models
```

```python
print('classification report: \n ', classification_report(y_test,y_pred))
```
